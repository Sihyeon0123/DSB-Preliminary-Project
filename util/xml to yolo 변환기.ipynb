```python
import xml.etree.ElementTree as ET
import os 

dir_path = './a'
output_dir_path = './b'
file_lst = os.listdir(dir_path)

for file in file_lst:
    xml_path = os.path.join(dir_path, file)

    # XML 파일 파싱
    tree = ET.parse(xml_path)
    root = tree.getroot()

    # 이미지 크기 정보 추출
    size = root.find('size')
    width = int(size.find('width').text)
    height = int(size.find('height').text)

    # 객체 정보 추출
    objects = root.findall('object')
    yolo_lines = []

    for obj in objects:
        name = obj.find('name').text
        if name == 'nh':
            class_id = 2  # fire는 0
        elif name == 'belt':
            class_id = 1  # smoke는 1
        else:
            continue  # 다른 클래스는 건너뜁니다

        bndbox = obj.find('bndbox')
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)

        # YOLO 형식의 바운딩 박스 좌표 계산
        x_center = (xmin + xmax) / 2.0 / width
        y_center = (ymin + ymax) / 2.0 / height
        bbox_width = (xmax - xmin) / width
        bbox_height = (ymax - ymin) / height

        # YOLO 형식 라인 만들기
        yolo_line = f"{class_id} {x_center:.6f} {y_center:.6f} {bbox_width:.6f} {bbox_height:.6f}"
        yolo_lines.append(yolo_line)

    output_file = os.path.join(output_dir_path, file.split('.')[0])
    output_file = output_file + '.txt'
    # YOLO 형식 파일로 저장
    with open(output_file, 'w') as f:
        f.write("\n".join(yolo_lines))

    print(f'YOLO 형식으로 변환된 라벨을 {output_file}에 저장했습니다.')
```
