# Deep Learning - LSTM
# 라이브러리 import

```python
import warnings
# 특정 UserWarning을 무시
warnings.filterwarnings("ignore", category=UserWarning)
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

# 데이터셋 로드

```python
df = pd.read_excel('../t.xlsx')
```

```python
# 라벨과 피처 분리
labels = df['label']
features = df.drop(columns=['label'])
```

```python
# 데이터 시퀀스 형태로 변환
num_frames = 5
num_features = 7  # 프레임당 각도 데이터의 수
X = features.values.reshape(-1, num_frames, num_features)
y = np.array(labels)
```

```python
# 훈련 및 테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)
```

```python
# LSTM 모델 정의
model = Sequential()
model.add(LSTM(50, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax'))  # 다중분류라면 Dense의 유닛 수와 활성화 함수 조정 필요

# 모델 컴파일   
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 모델 학습_
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```

```python
# 모델 예측
test_pred = model.predict(X_test)
test_pred_classes = np.argmax(test_pred, axis=1)

# 검증 세트 정확도
print("검증 세트 정확도:", accuracy_score(y_test, test_pred_classes))

# 검증 세트 보고서
print("검증 세트 보고서:\n", classification_report(y_test, test_pred_classes))
```
