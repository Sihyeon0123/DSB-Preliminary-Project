# 버전별 results.csv 합치기

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
v5_df = pd.read_csv('v5 results.csv')
v6_df = pd.read_csv('v6 results.csv')
v8_df = pd.read_csv('v8 results.csv')
```

```python
# metrics/mAP50(B) 속성 값만 추출
v5_map50 = v5_df[['metrics/mAP50(B)']].rename(columns={'metrics/mAP50(B)': 'v5 mAP50(B)'})
v6_map50 = v6_df[['metrics/mAP50(B)']].rename(columns={'metrics/mAP50(B)': 'v6 mAP50(B)'})
v8_map50 = v8_df[['metrics/mAP50(B)']].rename(columns={'metrics/mAP50(B)': 'v8 mAP50(B)'})
```

```python
# 각 데이터프레임을 하나로 결합
mAP50_df = pd.concat([v5_map50, v6_map50, v8_map50], axis=1)
mAP50_df.to_csv('YOLOv5_6_8_mAP50.csv', index=None)
mAP50_df.head()
```

```python
# metrics/mAP50(B) 속성 값만 추출
v5_map95 = v5_df[['metrics/mAP50-95(B)']].rename(columns={'metrics/mAP50-95(B)': 'v5 mAP95(B)'})
v6_map95 = v6_df[['metrics/mAP50-95(B)']].rename(columns={'metrics/mAP50-95(B)': 'v6 mAP95(B)'})
v8_map95 = v8_df[['metrics/mAP50-95(B)']].rename(columns={'metrics/mAP50-95(B)': 'v8 mAP95(B)'})
```

```python
# 각 데이터프레임을 하나로 결합
mAP95_df = pd.concat([v5_map95, v6_map95, v8_map95], axis=1)
mAP95_df.to_csv('YOLOv5_6_8_mAP95.csv')
mAP95_df.head()
```

# 그래프 그리기

```python
# 꺾은선 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(mAP95_df.index, mAP95_df['v5 mAP95(B)'], label='YOLOv5')
plt.plot(mAP95_df.index, mAP95_df['v6 mAP95(B)'], label='YOLOv6')
plt.plot(mAP95_df.index, mAP95_df['v8 mAP95(B)'], label='YOLOv8')
#
# 그래프 꾸미기
plt.xlabel('Epoch')
plt.ylabel('mAP50-95')
plt.title('mAP50-95 over Epochs')
plt.legend()
plt.grid(False)

# 그래프 보여주기
plt.show()
```

```python
# 꺾은선 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(mAP50_df.index, mAP50_df['v5 mAP50(B)'], label='YOLOv5')
plt.plot(mAP50_df.index, mAP50_df['v6 mAP50(B)'], label='YOLOv6')
plt.plot(mAP50_df.index, mAP50_df['v8 mAP50(B)'], label='YOLOv8')

# 그래프 꾸미기
plt.xlabel('Epoch')
plt.ylabel('mAP50')
plt.title('mAP50 over Epochs')
plt.legend()
plt.grid(False)

# 그래프 보여주기
plt.show()
```
