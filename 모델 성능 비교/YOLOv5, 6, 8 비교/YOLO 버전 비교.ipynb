```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('./mAP95.xlsx')
```

```python
df.head()
# yolov5, 6, 8
```

```python
# 꺾은선 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['yolov5'], label='YOLOv5')
plt.plot(df.index, df['yolov6'], label='YOLOv6')
plt.plot(df.index, df['yolov8'], label='YOLOv8')

# 그래프 꾸미기
plt.xlabel('Epoch')
plt.ylabel('mAP95')
plt.title('mAP95 over Epochs')
plt.legend()
plt.grid(True)

# 그래프 보여주기
plt.show()
```
